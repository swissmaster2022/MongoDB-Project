## Extract Load Transformation (ELT) Process

An **Extract Load Transformation  (ELT)** will be performed prior to analyzing the data. A quick overview includes the following:
- Extract the data (json file) from a URL from a website using requests and save to local json files.
- Load the json files into different collections in a local MongoDB database
- Transform the data using both Pymongo and Python libraries using MongoDB aggregate pipelines
- Analyze the data and answer the questions above.

#### Extract only 2015 Airline delay data

```python
#Aggregate to create 2015 only data in new collection "2015airlines"
# Unwind alldelays based on year 2015 and create new collection

mycol1 = mydb["alldelays"]
mycol2 = mydb["allairports"]

# Deconstruct array by Time, duplicating each document in the pipeline once per array element
unwind1 = { "$unwind": { "path": "$Time" }}

# Find only documents with year 2015
match1 =  {"$match":{"Time.Year": {"$in" :[2015]}}}

# Sort by the Airport code
sort1 = {"$sort": {"Airport.Code":1}}

# Create a collection in MongoDB to store the results
out1 = {"$out" : "2015airlines"}


# Limit results to 2
limit1 = {'$limit': 2}

# Execute the pipeline (out has been suppressed so as to display output)

pipeline1 =[unwind1, match1, sort1, limit1]

print("Pipeline successfully executed ...")

results1 = mycol1.aggregate(pipeline1)

df1 =pd.DataFrame(results1)

# Display Header

df2 = (df1[['Airport', 'Statistics', 'Time']])

# Display only the name, code, city and state
print("Sample of collection ..")

print(df2.head())
```

### Normalize - Percentage

In order to accurately report the correct overall value when comparing all airports, it is essential to "normalize" the data as some airports have less flights incoming and if using only Total Flights Delayed, we cannot accurately formulate an answer.  In this case, standardization has been achieved by using the total flights against the flights delayed to obtain a percentage value. This will bring all variables to the same level and the top 5 airlines with the highest delays will be more accurately calculated and displayed. Pandas dataframes will be used to display the data.


### Top 5 airports with the largest number of delays



```python
### Total Delays for 2015 

uscol = mydb["2015airlines"]
ustop5code =mydb["airportdelay-pct"]

limit_5 = {'$limit': 5}
out_air = {"$out" : "airportdelay-pct"} # create new collection from pipeline

sort_pct = {'$sort':{'Percentage': -1, "Airport.Code":1}}

sort = {'$sort':{
    'Year': 1,'Month': 1}}

group = {"$group": {
    "_id":"$Airport.Code", 
         "TotalDelays" : {"$sum" : "$Statistics.Flights.Delayed"},
        "TotalFlights" : {"$sum" : "$Statistics.Flights.Total"}}}
        
project = { "$project" : {'_id':1,
      'Total Delays' : '$TotalDelays',
      'Total Flights' : '$TotalFlights',
      "Percentage": {"$multiply":[{"$divide":["$TotalDelays","$TotalFlights"]},100]}
      },
     }
     

project_r = {"$project":{
                 "Percentage Delayed": { "$round": [ "$Percentage"] }}}         

# Execute pipeline, out_air which creates a collection has been omitted but will be used in the step below

pipeline4 =[group, project, sort_pct, limit_5]
         
for d in uscol.aggregate(pipeline4):
         dd = d
        
cursor = ustop5code.find()
print('Type of cursor:',type(cursor))        

# Converting cursor to the list of dictionaries
list_cur = list(cursor)

# Converting to the DataFrame
df1 = pd.DataFrame(list_cur)
print('Type of df:',type(df1))

#print(df1.head())
display(df1)

# printing color scales
#print(plotly_colorscales)

df1.plot(x="_id", y="Percentage", color='darkred', kind='bar', figsize=(16,6), title= 'US Airline Delays by Airports in 2015')
plt.xlabel('US Airports')
plt.ylabel('Delay Percentage')
plt.show()
```

```python
# Function in python that will read each string in the file and count the number of times it occurs.
# We use it to find the known airline carrier names

def word_count(str):
    # Create an empty dictionary
    counts = dict()
    words = str.split()

    # Loop through each line of the file
    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1
    return counts

# Open the file in read mode
file = open("carrier-clean.csv", "r")

#read content of file to string
data = file.read()

# Print the number of occurrences of word
print( word_count(data))
```
